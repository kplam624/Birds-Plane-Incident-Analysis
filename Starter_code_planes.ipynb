```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'Resources/database.csv'
df = pd.read_csv(file)
df.head()
```

```python
planes_df = df.loc[(df['Operator'] != 'UNKNOWN')]
planes_df
```

```python
df['Species Name'].value_counts()
```

```python
birds_df = df.loc[(df['Species Name'] != 'UNKNOWN BIRD') &
                  (df['Species Name'] != 'UNKNOWN MEDIUM BIRD') &
                  (df['Species Name'] != 'UNKNOWN SMALL BIRD') &
                  (df['Species Name'] != 'UNKNOWN TERRESTIAL BIRD') &
                  (df['Species Name'] != 'UNKNOWN LARGE BIRD')
                 ]
```

```python
birds_df
```

```python
bird_group = birds_df.groupby('Species Name').count()
```

```python
bird_sort = bird_group.sort_values(['Record ID'], ascending = False)
```

```python
bird_clean = bird_sort.nlargest(10,'Record ID')
bird_clean
```

```python
bird_names = ['MOURNING DOVE','GULL','KILLDEER','AMERICAN KESTREL','BARN SWALLOW','HORNED LARK',
              'EUROPEAN STARLING','SPARROW','ROCK PIGEON','RED-TAILED HAWK']
x_axis = np.arange(0,len(bird_clean))
birds = bird_clean['Record ID']
plt.bar(x_axis, birds , width = .5, color = 'blue', alpha = 1 ,align = 'center')
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, bird_names,rotation = 'vertical')

plt.figsize = (10,6)
plt.xlabel('Species Name')
plt.ylabel('Number of Incidents')
plt.title('Top 10 Birds of Death')
plt.tight_layout()

plt.savefig('output/birds_bar')
plt.show()
```

```python
label = bird_names
sizes = birds
colors = ['gold','aqua','darkmagenta','coral','silver','lawngreen',
         'steelblue','khaki','maroon','pink']
plt.pie(sizes,labels = label,colors = colors,autopct = '%.2f')

plt.savefig('output/birds_pie')
plt.show()
```
